```python
import pandas as pd

def display_parquet_sample(file_path, n_rows=405):
    """
    Reads a Parquet file and displays the first n_rows of data.
    
    Parameters:
        file_path (str): Path to the Parquet file.
        n_rows (int): Number of rows to display (default is 5).
    """
    try:
        df = pd.read_parquet(file_path)
        display(df.head(n_rows))
    except Exception as e:
        print(f"Failed to read the Parquet file: {e}")

# Update the file_path with your Parquet file's location
file_path = r"C:\Users\trege\OneDrive\Desktop\P8\RT1\Data_results_0001\Data\chunk-000\episode_000000.parquet"  # e.g., r"C:\Users\trege\output\data\chunk-000\episode_000001.parquet"

display_parquet_sample(file_path)
```
